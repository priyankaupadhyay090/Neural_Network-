```python
## # linespace function generates an array from start(-20) and stop(20) scale value
a = np.linspace(-20,20) # if you don't give samplevalue -it gives default 50
## or (-20, 20, 200), np.linspace(start, stop, num=50)  50 = number of samples generate between -20, 20 range
```

## a. Sigmoid Function



```python
## implementing Sigmoid Function and it's derivative with numpy and plot it with matplotlib on doamin [-20, 20]


def sigmoid(x):

    sig = 1/ (1 + np.exp(-x))
    
    return sig

def sigmoid_derivative(x):
    
    sig = sigmoid(x)
    sig_deriv_ds = sig * (1-sig)
    
    return sig_deriv_ds




## plot sigmoid and sigmoid_derivative with defined color combination

plt.plot(a, sigmoid(a), "green", label="Sigmoid")
plt.plot(a, sigmoid_derivative(a), "blue", label= "Sigmoid Derivative")
plt.legend(loc="upper right", frameon=False)
  
  
## X,Y plot with Title 

plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.title("Sigmoid Function, Sigmoid Derivative Function")

## plot with grid

plt.grid()

#show output

plt.show()




```

## b. ReLU (Rectified Linear Unit) Function 

```python
## implementing ReLU (Rectified Linear Unit) Function and it's derivative with numpy and plot it with matplotlib on doamin [-20, 20]



def ReLU(x):
    
    return x * (x > 0)
    
def ReLU_derivaive(x):
    
    return 1 * (x > 0)
    


## plot ReLU and ReLU_derivative with defined color combination and show

plt.plot(a, ReLU(a), "green",label="ReLU" )
plt.plot(a, ReLU_derivaive(a), "blue", label="ReLU Derivative")
plt.legend(loc="upper right", frameon=False)
    
  
## X,Y plot with Title

plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.title("ReLU Function, ReLU Derivaive Function")


## plot with grid

plt.grid()

#show output

plt.show()



```

## c. Tanh Function

```python
## implementing Tanh Function and it's derivative with numpy and plot it with matplotlib on doamin [-20, 20]



def Tanh(x):
    t = np.tanh(x)
    return t
    
def Tanh_derivative(x):
    t = Tanh(x)
    derivative_dt= 1-t**2
    return derivative_dt
    


## plot Tanh and Tanh_derivative with defined color combination and show

plt.plot(a, Tanh(a), "green", label="Tanh")
plt.plot(a, Tanh_derivative(a), "blue", label="Tanh derivative")
plt.legend(loc="upper right", frameon=False)
   
  
## X,Y plot with Title 

plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.title("Tanh Function, Tanh Derivaive Function")


## plot with grid

plt.grid()

#show output

plt.show()

```
